```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42) # generate same random numbers
```

```python
X = np.random.rand(100, 1) - 0.5
```

```python
y = 3 * X[:, 0] ** 2 - 0.05 * np.random.randn(100)
```

```python
df = pd.DataFrame(X, columns=["X"])
df['y'] = y
df.head()
```

```python
plt.scatter(df["X"], df["y"])
plt.show()
```

```python
df["pred1"] = df['y'].mean()
```

```python
df
```

```python
df["res1"] = df["y"] - df["pred1"]
```

```python
df
```

```python
plt.scatter(df['X'], df['y'])
plt.plot(df['X'], df['pred1'], color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
tree1 = DecisionTreeRegressor(max_leaf_nodes=8) # max_leaf_nodes value from 8 to 32 gives better results, statistically proven
tree1.fit(df['X'].values.reshape(100, 1), df['res1'].values)
```

```python
df["res1"].values.shape
```

```python
plot_tree(tree1)
plt.show()
```

```python
# generating X_test
X_test = np.linspace(-0.5, 0.5, 500)
X_test
```

```python
y_pred = 0.265566 + tree1.predict(X_test.reshape(500, 1))
```

```python
y_pred
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color='red')
plt.scatter(df["X"], df["y"])
plt.show()
```

```python
df['pred2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))
df
```

```python
df['res2'] = df['y'] - df['pred2']
df
```

```python
tree2 = DecisionTreeRegressor(max_leaf_nodes=8)
tree2.fit(df["X"].values.reshape(100, 1), df["res2"].values)
```

```python
plot_tree(tree2)
```

```python
X_test.reshape(-1, 1).shape
```

```python
final_pred = 0.265566 + sum(regressor.predict(X_test.reshape(-1, 1)) for regressor in [tree1, tree2])
```

```python
final_pred
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
plt.title('X vs y')
```

```python
def gradient_boost(X,y,number,lr,count=1,regs=[],foo=None):

  if number == 0:
    return 
  else:
    # do gradient boosting

    if count > 1:
      y = y - regs[-1].predict(X)
    else:
      foo = y

    tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
    tree_reg.fit(X, y)

    regs.append(tree_reg)

    x1 = np.linspace(-0.5, 0.5, 500)
    y_pred = sum(lr * regressor.predict(x1.reshape(-1, 1)) for regressor in regs)

    print(number)
    plt.figure()
    plt.plot(x1, y_pred, linewidth=2)
    plt.plot(X[:, 0], foo,"r.")
    plt.show()

    gradient_boost(X,y,number-1,lr,count+1,regs,foo=foo)
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
gradient_boost(X,y,5,lr=1)
```
