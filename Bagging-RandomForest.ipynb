```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
X, y = make_classification(n_features=5, n_informative=1, n_redundant=0, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X, columns=["col1", "col2", "col3", "col4", "col5"])
df["target"] = y
df
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:, :5], df.iloc[:, -1])
```

```python
plt.figure(figsize=(12, 12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rafc = RandomForestClassifier(max_features=2)
```

```python
rafc.fit(df.iloc[:, :5], df.iloc[:, -1])
```

```python
plt.figure(figsize=(12, 12))
plot_tree(rafc.estimators_[6])
plt.show()
```

#### As we can see from the two trees above, the first one is made from BaggingClassifier and in this algorithms, every node is determined from either of the one rows that we've chosen earlier. Therefore, there is node level sampling
